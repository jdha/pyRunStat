## An Example notebook on the use of pyRunStat

Using the pyRunStat python package plots of the basic statistics from a NEMO simulation can be plotted.

```python
import pyrunstat.extract as rsextract
import pyrunstat.plot as rsplot
```

### Plotting from the run.stat text file

```python
cd /Users/jdha/Projects/GitHub/pyRunStat
```

```python
file_path = './examples/data/run.stat.simA'

# Extract the 'ssh_max'
da_ssh_max = rsextract.read_text(file_path, 'U_max')

# Plot ssh_max timeseries
rsplot.time_series(da_ssh_max, labels=['NEMO sim A'])
```

### Plotting from the run.stat.nc netcdf file

```python
file_path = './run.stat.nc'

# Extract the 'ssh_max'
da_ssh_max = rs.extract.read_nc(file_path, 'ssh_max')

# Plot ssh_max timeseries
rs.plot.time_series(da_ssh_max, labels=['SSH maximum'])
```

### Plotting multiple simulations

```python
file_paths = ['./examples/data/run.stat.simA', 
              './examples/data/run.stat.simB', 
              './examples/data/run.stat.simC']
nfiles = len(file_paths)
da_U_max = [None] * len(file_paths)

# Extract the 'U_max'
for n in range(nfiles):
    da_U_max[n] = rsextract.read_text(file_paths[n], 'U_max')

# Plot U_max timeseries
rsplot.time_series(*da_U_max, labels=['NEMO sim A', 
                                      'NEMO sim B', 
                                      'NEMO sim C'])
```
